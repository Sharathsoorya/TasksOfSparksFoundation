```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
plt.figure(figsize=(8,4))
sns.scatterplot(data = df, s = 100, alpha = 0.7)
plt.grid()
plt.show()
```

```
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```
x.head()
```

```
y.head()
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

```
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = "entropy",  max_depth = 4)
classifier.fit(x_train, y_train)
```

```
y_pred = classifier.predict(x_test)
y_pred
```

```
from sklearn import metrics
metrics.accuracy_score(y_test, y_pred)
```

```
from sklearn.metrics import confusion_matrix
matrix = confusion_matrix(y_test, y_pred)
matrix
```

```
import pydotplus
```

```
from io import StringIO
import matplotlib.image as mimg
from sklearn import tree
from sklearn.tree import export_graphviz
```

```
text = tree.export_text(classifier)
text
```

```
dot_data = StringIO()
filename = "iris_test.jpg"
features = df.columns.tolist()[:-1]
target = df['Species'].unique()
fig = plt.figure(figsize=(15, 15))
dt = tree.export_graphviz(classifier, feature_names = features, out_file=dot_data, class_names = target,
                          filled = True, rotate = False, rounded = True, special_characters = True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png(filename)
img = mimg.imread(filename)
plt.figure(figsize=(100, 100),dpi = 100)
plt.imshow(img, interpolation = 'nearest')
```
